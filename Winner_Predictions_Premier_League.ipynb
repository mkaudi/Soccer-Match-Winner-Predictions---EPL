```python
# Create separate columns for 'home' and 'away'
def home_away_cols(df):
    '''
    The function separates the home and away scores.
    Formats the referee's name and then drops both the original score and referee columns


    args: 
        df - the dataframe
        
    returns:
        dataframe with the new formatted columns.
    
    '''

    df['home_score'] = df['score_fulltime'].apply(lambda x: x['home'])
    df['away_score'] = df['score_fulltime'].apply(lambda x: x['away'])
    df['referee_name'] = df['referee'].apply(lambda x: [item['name'] for item in x])
    df['referee_name'] = df['referee_name'].apply(lambda x: ','.join(x))

    df = df.drop(columns=['referee','score_fulltime'])

    return df

df= home_away_cols(df)
df_2023 =home_away_cols(df_2023)
```

```python
#team playing home's home win streak

def streak(df):
    '''
    Calculates the number of games the home team has won playing home before the current game.

    args: 
        df - the dataframe
        
    returns:
        dataframe with the new winning streak column.
    
    '''

    df_streaks = df[df['status_mapped'] == 1]
    
    win_streak = 0
    loss_streak = 0
    
    count = 0
    
    teams = df['home_team'].unique()
    
    for i in range(df['home_team'].nunique()):
        #create a dataframe for a team
        team_df = df[df['home_team'] == teams[i]]
    
        #iterate through the rows of the team dataframe
        for index, row in team_df.iterrows():
            #if the column name doesnt exist, create it and set it to 0
            if 'home_win_streak' not in team_df.columns:
                #team_df['home_win_streak'] = 0 
                team_df.at[index, 'home_win_streak'] = 0
                
            else:
                team_df.at[index, 'home_win_streak'] = win_streak
                #if the columns exists, and the winner of the match is the home team, 
                #set the win_streak to the previous value and increment the win_streak counter
                if row['winner_mapped'] == 1.0:
                    win_streak +=1
                    loss_streak = 0
    
                else:
                    win_streak = 0
                    loss_streak +=1
        
            df = pd.merge(df, team_df[['date','home_team', 'away_team', 'home_win_streak']], 
                          on = ['date','home_team', 'away_team'], how='left', suffixes=('', '_team') )
        
    # Filter columns that end with '_team'
    team_columns = df.filter(like='_team')

    # Convert all columns to numeric, setting errors='coerce' will replace non-numeric values with NaN
    team_columns = team_columns.apply(pd.to_numeric, errors='coerce')

    # Fill NaN values with a default value, e.g., 0
    #team_columns = team_columns.fillna(0)
    
    df['home_win_streak'] = team_columns.iloc[:, 2:].max(axis=1)
    df = df.drop(columns=team_columns.iloc[:, 2:])

    return df
```

```python
#home average goals when playing at home

def avg_goals(df):
    '''
    Calculates the average number of goals the home team has scored in the previous games before the current game.

    args: 
        df - the dataframe
        
    returns:
        dataframe with the new column.
    
    '''

    df_streaks = df[df['status_mapped'] == 1]
        
    #win_streak = 0
    #loss_streak = 0
    
    game_count = 0
    goals_count = 0
    
    teams = df["home_team"].unique()
    
    for i in range(df['home_team'].nunique()):
        #create a dataframe for a team
        team_gl_df = df[df['home_team'] == teams[i]]
    
        #iterate through the rows of the team dataframe
        for index, row in team_gl_df.iterrows():
            #if the column name doesnt exist, create it and set it to 0
            if 'home_avg_goals_when_home' not in team_gl_df.columns:
                team_gl_df.at[index, 'home_avg_goals_when_home'] = team_gl_df.at[index, 'home_score']
            else:
                if game_count == 0:
                    team_gl_df.at[index, 'home_avg_goals_when_home'] = 0
                else:
                    team_gl_df.at[index, 'home_avg_goals_when_home'] = goals_count/game_count
                
                #print(goals_count)
                #print(game_count)
                goals_count += team_gl_df.at[index, 'home_score']
                game_count +=1
    
            df = pd.merge(df, team_gl_df[['date','home_team', 'away_team', 'home_avg_goals_when_home']], 
                          on = ['date','home_team', 'away_team'], how='left', suffixes=('', '_team') )
            
    # Filter columns that end with '_team'
    team_columns = df.filter(like='_team')
    
    # Convert all columns to numeric, setting errors='coerce' will replace non-numeric values with NaN
    team_columns = team_columns.apply(pd.to_numeric, errors='coerce')

    # Fill NaN values with a default value, e.g., 0
    
    df['home_avg_goals_when_home'] = team_columns.iloc[:, 2:].max(axis=1)
    df = df.drop(columns=team_columns.iloc[:, 2:])
    df = df.fillna(0)

    return df
```

```python
#away average goals when playing away

def away_avg_goals(df):
    '''
    Calculates the average number of goals the away team has scored when playing away

    args: 
        df - the dataframe
        
    returns:
        dataframe with the new column.
    '''
    
    df_streaks = df[df['status_mapped'] == 1]
        
    #win_streak = 0
    #loss_streak = 0
    
    game_count = 0
    goals_count = 0
    
    teams = df["away_team"].unique()
    
    for i in range(df['away_team'].nunique()):
        #create a dataframe for a team
        team_gl_df = df[df['away_team'] == teams[i]]
    
        #iterate through the rows of the team dataframe
        for index, row in team_gl_df.iterrows():
            #if the column name doesnt exist, create it and set it to 0
            if 'away_avg_goals_when_away' not in team_gl_df.columns:
                team_gl_df.at[index, 'away_avg_goals_when_away'] = team_gl_df.at[index, 'away_score']
            else:
                if game_count == 0:
                    team_gl_df.at[index, 'away_avg_goals_when_away'] = 0
                else:
                    team_gl_df.at[index, 'away_avg_goals_when_away'] = goals_count/game_count
                
                #print(goals_count)
                #print(game_count)
                goals_count += team_gl_df.at[index, 'away_score']
                game_count +=1
    
            df = pd.merge(df, team_gl_df[['date','home_team', 'away_team', 'away_avg_goals_when_away']], 
                          on = ['date','home_team', 'away_team'], how='left', suffixes=('', '_team') )
            
    # Filter columns that end with '_team'
    team_columns = df.filter(like='_team')
    team_columns = team_columns.apply(pd.to_numeric, errors='coerce')

    #print(team_columns)
    df['away_avg_goals_when_away'] = team_columns.iloc[:, 2:].max(axis=1)
    df = df.drop(columns=team_columns.iloc[:, 2:])
    df = df.fillna(0)

    return df
```

```python

def head2head(df):
    '''
    Identify which team won the last time the 2 teams met.

    args: 
        df - the dataframe
        
    returns:
        dataframe with the new column.
    '''
    
    df['team_names_combined'] = df['home_team'] + df['away_team']
    df['head_to_head'] = 0
    
    def sort_letters(text):
        return ''.join(sorted(text))
    
    for i in range(1, len(df)):
        current_text_sorted = sort_letters(df.at[i, 'team_names_combined'])
        # Search upwards for the same combination of letters
        for j in range(i):
            if sort_letters(df.at[j, 'team_names_combined']) == current_text_sorted:
                df.at[i, 'head_to_head'] = df.at[j, 'winner_mapped']
                break
    
    df = df.drop(columns=['team_names_combined'])
    
    return df

```

```python
def predict_multi(predict_data, model):
    '''
    The function performs prdictions on data provided using the provided trained model.

    args:
        predict_data: the data to predict on.
        model: the trained model to use for predicting.

    return:
        predict_data: the dataframe with the predictions and probabilities included.
    '''

    #predict on the data
    predictions = rf.predict(predict_data)

    #predict the probabilities for the different(3) classes
    predictions_proba = rf.predict_proba(predict_data)

    #the dataframe inclusive of the class predicitons
    pred_df = pd.DataFrame(predictions, columns=['winner_prediction'], index=predict_data.index)

    #the dataframe inclusive of the class probabilities predicitons
    prob_df = pd.DataFrame(predictions_proba, columns=['prob_class_0', 'prob_class_1','prob_class_2'], index=predict_data.index)

    #merge the winner predictions and the probailities into the prediction dataframe
    predict_data = pd.merge(predict_data, pred_df, left_index=True, right_index=True, how='left')
    predict_data = pd.merge(predict_data, prob_df, left_index=True, right_index=True, how='left')

    return predict_data
```
